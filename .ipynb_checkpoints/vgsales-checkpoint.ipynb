```python
def sales_through_years(df, Field, fig, ax, area = ("Global_Sales", "Global"), Con = "0", con = 0, start_year = 1980, 
                        end_year = 2018, top_count = 5, year_label = 0, legend_label = 1):
    
    a = df.loc[(df["Year"] >= start_year) & (df["Year"] < end_year) & (df[Con] == con)]
    g = a.groupby(Field)
    sales_area = g.apply(lambda a: a[area[0]].sum()).sort_values(ascending = False).iloc[:top_count]

    if (con != 0): con_str = "of " + con + " "
    else: con_str = ""    
    for field in sales_area.keys():
        g = df.groupby("Year")
        b = g.apply(lambda a: a[a[Field] == field][area[0]].sum())
        b = b[(b.index >= start_year) & (b.index < end_year)]
        ax.plot(b.keys(), b.values, label = field, color = plat_col[field])

    ax.set_title("Top " + str(top_count) + " " + Field + " for sales " + con_str 
                 + area[1] + " " + str(start_year) + " - " + str(end_year - 1))
    if (year_label): ax.set_xticks(range(start_year, end_year, 1))
    if (legend_label): ax.legend(title = Field, loc = "upper left", frameon = True)
    _ = ax.grid(visible = True, color='steelblue', linestyle='-.', linewidth=0.5, zorder = 0)
    
def top_sales(Field, fig, ax, area = ("Global_Sales", "Global"), Con = "0", con = 0,
              start_year = 1980, end_year = 2018, top_count = 10, full_label = 0, col = "darkslateblue"):
    
    a = vgs.loc[(vgs["Year"] >= start_year) & (vgs["Year"] < end_year) & (vgs[Con] == con)]
    g = a.groupby(Field)
    sales_area = g.apply(lambda a: a[area[0]].sum()).sort_values(ascending = False).iloc[:top_count]
    
    ax.bar(sales_area.keys(), sales_area.values, zorder = 3, color = col)
    if (con != 0): con_str = "of " + con + " "
    else: con_str = ""
    ax.set_title("Top " + str(top_count) + " " + Field + " for sales " + con_str 
                      + area[1] + " " + str(start_year) + " - " + str(end_year - 1))
    _ = ax.set_xticklabels(labels = shorten_label(sales_area, full_label).values())
    _ = ax.grid(visible = True, color='steelblue', linestyle='-.', linewidth=0.5, zorder = 0)
    
def shorten_label(sales_area, full_label = 0):
    short_labels = dict()
    for label in sales_area.keys():
        labels = str(label).split()
        short_labels[label] = ""
        count = 0
        for word in labels:
            short_labels[label] += (word + "\n")
            count += 1
            if ((not full_label) & (count == 2)): break
        
    return(short_labels)

def rainbow(n, s, v):
    rgb = list()

    for hue in np.arange(0, 0.8, 0.8/n):
        (r, g, b) = colorsys.hsv_to_rgb(hue, s, v)
        rgb.append([r,g,b])

    return rgb

sales_list = ["NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales", "Other_Sales"]
name_list = ["North America", "Europe", "Japan" , "Global", "the rest of the world"]
col_list = ["limegreen", "royalblue", "indianred"]

# rgb = mcolors.CSS4_COLORS.keys()
# plat_col = dict(zip(vgs["Platform"].value_counts().keys(), rgb))
plat_col = {
 'DS': 'b',
 'PS2': 'g',
 'PS3': 'r',
 'Wii': 'c',
 'X360': 'm',
 'PSP': 'y',
 'PS': 'k',
 'PC': 'purple',
 'XB': 'aqua',
 'GBA': 'yellow',
 'GC': 'orangered',
 '3DS': 'brown',
 'PSV': 'lime',
 'PS4': 'crimson',
 'N64': 'cornflowerblue',
 'SNES': 'indigo',
 'XOne': 'coral',
 'SAT': 'cornflowerblue',
 'WiiU': 'cornsilk',
 '2600': 'crimson',
 'NES': 'cyan',
 'GB': 'darkblue',
 'DC': 'darkcyan',
 'GEN': 'darkgoldenrod',
 'NG': 'darkgray',
 'SCD': 'darkgreen',
 'WS': 'darkgrey',
 '3DO': 'darkkhaki',
 'TG16': 'darkmagenta',
 'GG': 'darkolivegreen',
 'PCFX': 'darkorange'}

areas = list(zip(sales_list, name_list))
```

```python
name_count = vgs["Games"].value_counts()
name_other_count = np.sum(name_count[5:].values)
name_count = name_count.drop(name_count[5:].keys())

plat_count = vgs["Platform"].value_counts()
plat_other_count = np.sum(plat_count[5:].values)
plat_count = plat_count.drop(plat_count[5:].keys())
plat_count["Others"] = plat_other_count

year_count = vgs["Year"].value_counts().sort_index()

genre_count = vgs["Genre"].value_counts()
genre_other_count = np.sum(genre_count[12:].values)
genre_count = genre_count.drop(genre_count[12:].keys())

publisher_count = vgs["Publisher"].value_counts()
publisher_count = publisher_count.drop(publisher_count[5:].keys())

fig = plt.figure(figsize = (15,6))
grid = plt.GridSpec(2,3, hspace = 0.5, wspace = 0.2)
ax_publisher = fig.add_subplot(grid[0,0])
ax_plat = fig.add_subplot(grid[0,1])
ax_year = fig.add_subplot(grid[0,2])
ax_genre = fig.add_subplot(grid[1,1:])
ax_genre.set_xticklabels(genre_count.keys(),rotation = -45)
ax_name = fig.add_subplot(grid[1,0])

ax_name.barh(name_count.keys(), name_count.values, color = "darkslateblue")
ax_plat.bar(plat_count.keys(), plat_count.values, color = "limegreen")
ax_year.plot(year_count.keys(), year_count.values, color = "gold")
ax_genre.bar(genre_count.keys(), genre_count.values, color = rainbow(12, 0.63, 0.87)
             , edgecolor = rainbow(12, 1, 0.63), linewidth = 1.3)
ax_publisher.barh(publisher_count.keys(),publisher_count.values, color = "salmon")

ax_publisher.set_title("Top 5 publisher for releases")
ax_plat.set_title("Top 5 platform for releases")
ax_year.set_title("Releases in 1980 - 2020")
ax_genre.set_title("Top 12 genre for releases")
ax_name.set_title("Top 5 games for releases")
```

```python
df = pd.DataFrame()
df.index = ['Count', 'Mean', 'Min', '25%', '50%', '75%', 'Max', 'Std']

for col in ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]:
    count = vgs[col].shape[0]
    mean = vgs[col].mean()
    col_min = np.min(vgs[col])
    p_25 = vgs[col].quantile(.25)
    p_50 = vgs[col].quantile(.5)
    p_75 = vgs[col].quantile(.75)
    col_max = np.max(vgs[col])
    std = vgs[col].std()
    new_col = [count, mean, col_min, p_25, p_50, p_75, col_max, std]
    df[col] = new_col
df
    
```

```python
fig, axes = plt.subplots(nrows = 6, figsize = (15,18))
fig.subplots_adjust(hspace = 0.5)

g = vgs.groupby("Year")
a = g.apply(lambda x: x["Global_Sales"].sum()).sort_values(ascending = False).iloc[:10].sort_index()
year_label = [str(int(u)) for u in a.keys()]

color = list()
edgecolor = list()
amax = max(a.values)
for i in a.values:
    (r,g,b) = colorsys.hsv_to_rgb(1 - (i + 1000) / (amax + 1000), 0.63, 0.87)
    color.append([r,g,b])
    (r,g,b) = colorsys.hsv_to_rgb(1 - (i + 1000) / (amax + 1000), 1, 0.6)
    edgecolor.append([r,g,b])
    
axes[0].bar(year_label, a.values, zorder = 3, color = color, edgecolor = edgecolor, linewidth = 1.5)
axes[0].set_title("Top 10 years for sales Global 1980 - 2017")
_ = axes[0].grid(visible = True, color='steelblue', linestyle='-.', linewidth=0.5, zorder = 0)

rgb = rainbow(5, 0.63, 0.87)

pub_rel = vgs["Publisher"].value_counts()
pub_rel = pub_rel.drop(pub_rel[10:].keys())
axes[1].bar(pub_rel.keys(), pub_rel.values, zorder = 3, color = rgb[0])
_ = axes[1].set_xticklabels(shorten_label(pub_rel).values())
_ = axes[1].set_title("Top 10 Publisher for releases in Global 1980 - 2017")
_ = axes[1].grid(visible = True, color='steelblue', linestyle='-.', linewidth=0.5, zorder = 0)

top_sales(Field = "Publisher", fig = fig, ax = axes[2], col = rgb[1])
top_sales(Field = "Platform", fig = fig, ax = axes[3], col = rgb[2])
top_sales(Field = "Genre", fig = fig, ax = axes[4], col = rgb[3])
top_sales(Field = "Games", fig = fig, ax = axes[5], col = rgb[4], full_label = True)
```
